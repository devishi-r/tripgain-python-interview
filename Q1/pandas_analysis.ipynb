```python
import pandas as pd
df = pd.read_csv("data.csv")
```

```python
print(len(df))
print(df.columns)
print(df.head())
print(df.describe())
```

```python
close_matches = df[(df["win_by_runs"]==1) | (df["win_by_wickets"]==1)]
print(close_matches["player_of_match"].value_counts().idxmax())
```

```python
wkhd = df[df["venue"]=="Wankhede Stadium"]
runs = wkhd[wkhd["win_by_runs"]>0].shape[0]
bat = wkhd[wkhd["win_by_wickets"]>0].shape[0]
# print(runs)
# print(bat)
if(runs>bat):
    print("By wicket")
else:
    print("By runs")
# wkhd.head()
```

```python
wins = df[df["win_by_runs"]>50]
print(wins["winner"].value_counts().idxmax())
```

```python
wtoss = df[df["toss_winner"]==df["winner"]]
print(len(wtoss))
```

```python
kkr = df[(df["team1"]=="Kolkata Knight Riders") | (df["team2"]=="Kolkata Knight Riders")]
ump1 = kkr["umpire1"].value_counts().sum()
ump2 = kkr["umpire2"].value_counts().sum()
# print(ump1, ump2)
if(ump1>ump2):
    print("Umpire 1")
else:
    print("Umpire 2")
```
